```python
import pandas as pd
import numpy as np
from sklearn.metrics import accuracy_score, precision_score, recall_score
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.neighbors import KNeighborsClassifier
```

```python
df = pd.read_csv("IRIS.csv")
```

```python
df.head()
```

```python
df = df.drop("Id", axis=1)
```

```python
df.head()
```

```python
x = df.drop("Species", axis=1)
y = df["Species"]

x_train, x_test, y_train, y_test = train_test_split(
    x, y, test_size = 0.5, random_state = 42
)

scaler = StandardScaler()
x_train_scaled = scaler.fit_transform(x_train)
x_test_scaled = scaler.transform(x_test)

knn_model = KNeighborsClassifier(n_neighbors = 9)
knn_model.fit(x_train_scaled, y_train)

y_pred = knn_model.predict(x_test_scaled)

print("accuracy_score:", accuracy_score(y_test, y_pred) * 100, "%")
print("precision_score:", precision_score(y_test, y_pred, average='weighted') * 100, "%")
print("recall_score:", recall_score(y_test, y_pred, average='weighted') * 100, "%")
```

```python
from sklearn.model_selection import GridSearchCV

classifier = KNeighborsClassifier()
grid = {"n_neighbors":[3,5,7,9]}
classifierCV = GridSearchCV(
    classifier,
    grid,
    cv = 5
)

classifierCV.fit(x_train_scaled, y_train)

y_pred = classifierCV.predict(x_test_scaled)

print("accuracy_score:", accuracy_score(y_test, y_pred) * 100, "%")
print("precision_score:", precision_score(y_test, y_pred, average='weighted') * 100, "%")
print("recall_score:", recall_score(y_test, y_pred, average='weighted') * 100, "%")
```

```python
import pandas as pd

# result
res = pd.DataFrame(classifierCV.cv_results_)
print(res[["param_n_neighbors","mean_score_time"]])
```
